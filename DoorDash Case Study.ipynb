```python
df = pd.read_csv("Doordash v4.csv")
df.head()
```

```python
df.dropna(how='any', inplace=True)
df.info()
```

```python
df = df.rename(columns={'Is ASAP': 'immediate'})
```

```python
df.drop(columns=['app_gap', 'driver_gap', 'delivery'], inplace=True)
df.head()
```

```python
df['order_placed'] =  pd.to_datetime(df['order_placed'],format='%m/%d/%Y %H:%M:%S')
df['Restaurant_order'] =  pd.to_datetime(df['Restaurant_order'],format='%m/%d/%Y %H:%M:%S')
df['driver_arrived'] =  pd.to_datetime(df['driver_arrived'],format='%m/%d/%Y %H:%M:%S')
df['delivered'] =  pd.to_datetime(df['delivered'],format='%m/%d/%Y %H:%M:%S')
```

```python
df['doordash_gap'] = (df['Restaurant_order'] - df['order_placed'])
df['driver_gap'] = (df['driver_arrived'] - df['Restaurant_order'])
df['delivery_gap'] = (df['delivered'] - df['driver_arrived'])
df['total_time'] = (df['delivered'] - df['order_placed'])
df.head()
```
